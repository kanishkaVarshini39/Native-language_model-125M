```python
import torch
import tiktoken
from src.config import CONFIG_124M
from src.model import BaseModel
from src.utils import text_to_token_ids, token_ids_to_text, generate
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "mps" if torch.backends.mps.is_available() else "cpu")
print(f"Using device: {device}")

model = BaseModel(CONFIG_124M)
model.to(device)
tokenizer = tiktoken.encoding_for_model("gpt2")
```

```python
checkpoint = torch.load("model_and_optimizer.pth", map_location=device, weights_only=True)
model.load_state_dict(checkpoint["model_state_dict"])
model.eval()
print("Model weights loaded successfully.")
```

```python
torch.manual_seed(123)

start_text = "Every effort moves you"
encoded_input = text_to_token_ids(start_text, tokenizer).to(device)

token_ids = generate(
    model=model,
    idx=encoded_input,
    max_new_tokens=15,
    context_size=CONFIG_124M["context_length"],
    top_k=25,
    temperature=1.6
)

print("Input context:", start_text)
print("Output text:\n", token_ids_to_text(token_ids, tokenizer))
```
